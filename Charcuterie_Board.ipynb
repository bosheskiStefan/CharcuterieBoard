```python
import re

food_re=r'^[A-Z]{1}[a-z]*$'

charcuterie_board=[]
price=50

cheeses={
    "Swiss": 7.40,
    "Gorgonzola": 12.50,
    "Parmesan": 5.90,
    "Mozzarella": 3.80,
    "Cheddar": 8.85,
    "Gouda": 2.30,
    "Edam": 2.30,
    "White": 2.75
}
fruits={
    "Fig": 3.00,
    "Apricot": 3.90,
    "Grape": 4.60,
    "Apple": 2.40,
    "Berries": 4.50
}
meats={
    "Ham": 9.30,
    "Prosciutto": 12.20,
    "Pepperoni": 11.75,
    "Chicken": 14.90
}
extras={
    "Crackers": 3.70,
    "Bread": 2.50,
    "Honey": 8.50,
    "Jam": 7.45,
    "Butter": 6.10
}

def add_board(charcuterie_board, price):
  flag=0
  print("What do you want to add to your Charcuterie Board?")
  choice = input("(1) Cheese\n(2) Fruit\n(3) Meat\n(4) Extras:\n")
  choice = int(choice)
  if choice in (1, 2, 3, 4):
    if choice == 1:
      while True:
        for key, value in cheeses.items():
          print(f'{key}: $ {round(value, 2)}')
        cheese = input("Which cheese do you want to add?")
        cheese = str(cheese)
        if re.match(food_re, cheese):
          for key, value in cheeses.items():
            if key == cheese:
              charcuterie_board.append(key)
              price=price+float(value)
              flag=1
              break
          if flag == 0:
            print("Item not found!")
            continue
          break
        else:
          print("Wrong input!")
    elif choice == 2:
      while True:
        for key, value in fruits.items():
          print(f'{key}: $ {round(value, 2)}')
        fruit = input("Which fruit do you want to add?")
        fruit = str(fruit)
        if re.match(food_re, fruit):
          for key, value in fruits.items():
            if key == fruit:
              charcuterie_board.append(key)
              price=price+float(value)
              flag=1
              break
          if flag == 0:
            print("Item not found!")
            continue
          break
        else:
          print("Wrong input!")
    elif choice == 3:
      while True:
        for key, value in meats.items():
          print(f'{key}: $ {value}')
        meat = input("Which meat do you want to add?")
        meat = str(meat)
        if re.match(food_re, meat):
          for key, value in meats.items():
            if key == meat:
              charcuterie_board.append(key)
              price=price+float(value)
              flag=1
              break
          if flag == 0:
            print("Item not found!")
            continue
          break
        else:
          print("Wrong input!")
    elif choice == 4:
      while True:
        for key, value in extras.items():
          print(f'{key}: $ {value}')
        extra = input("Which extra thing do you want to add?")
        extra = str(extra)
        if re.match(food_re, extra):
          for key, value in extras.items():
            if key == extra:
              charcuterie_board.append(key)
              price=price+float(value)
              flag=1
              break
          if flag == 0:
            print("Item not found!")
            continue
          break
        else:
          print("Wrong input!")
  else:
    print("Wrong choice!")
  return price

def remove_board(charcuterie_board, price):
  print(charcuterie_board)
  choice = input("What do you want to remove from the board?")
  choice = str(choice)
  if re.match(food_re, choice):
    for stuff in charcuterie_board:
      if stuff == choice:
        charcuterie_board.remove(choice)
        if choice in cheeses:
          price=price-float(cheeses[choice])
        elif choice in fruits:
          price=price-float(fruits[choice])
        elif choice in meats:
          price=price-float(meats[choice])
        elif choice in extras:
          price=price-float(extras[choice])
  else:
    print("Wrong input!")
  return price

def vegetarian(charcuterie_board):
  flag=0
  for stuff in charcuterie_board:
    if stuff in meats:
      print("The charcuterie board is not vegetarian.")
      flag=1
  if flag==0:
    print("The charcuterie board is vegetarian.")

def print_board(charcuterie_board, price):
  print(f'$ {round(float(price), 2)}')
  print(tuple(charcuterie_board))

while True:
  print("Choose an option.")
  print("1. Adding to the board")
  print("2. Removing to the board")
  print("3. Show the board")
  print("4. Is the board vegetarian")
  print("0. Exiting")
  option=input()
  option=int(option)
  if option == 1:
    price=add_board(charcuterie_board, price)
  elif option == 2:
    price=remove_board(charcuterie_board, price)
  elif option == 3:
    print_board(charcuterie_board, price)
  elif option == 4:
    vegetarian(charcuterie_board)
  elif option == 0:
    break
  else:
    print("Wrong input!")
```
